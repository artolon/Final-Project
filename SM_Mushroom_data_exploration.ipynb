## Exploratory Data Analysis

### From Abbie's code

```python
from sklearn.preprocessing import LabelEncoder

mushroom_data = mushroom_data.astype('category')

labelencoder=LabelEncoder()
for c in mushroom_data.columns:
    mushroom_data[c] = labelencoder.fit_transform(mushroom_data[c])

mushroom_data.head()
```

```python
mushroom_data_m = pd.melt(mushroom_data,'class', var_name='Characteristics')
mushroom_data_m['class'].replace({0:'Edible', 1:'Poisonous'}, inplace=True)
fig, ax = plt.subplots(figsize=(20,8))
sns.violinplot(x='Characteristics', y='value', hue='class', split = True, data=mushroom_data_m, palette = 'prism')
features = mushroom_data.drop(['class'],axis = 1)
ax.set_xticklabels(rotation = 90, labels = list(features.columns), size=12)
plt.show()
```

##### Below features do not show significant value distribution across classes 
    * gill-attachment
    * gill-spacing
    * gill-size
    * veil-type
    * veil-color
    * ring-number
##### Hence we will drop these columns and then check correlation

```python
mushroom_features = mushroom_data.drop(['gill-attachment', 'gill-spacing', 'gill-size', 'veil-type', 'veil-color', 'ring-number'],axis=1)
mushroom_features.shape
```

```python
# Checking heatmap
fig, ax = plt.subplots(figsize=(20,10))

corr = mushroom_features.corr()
sns.heatmap(corr, annot=True)
plt.show()
```
